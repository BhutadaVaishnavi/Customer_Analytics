# 17. Find out the Ratio of browsing time w.r.t total purchases by location 

```python
L= cd.groupby(["Location"])["Product_Browsing_Time","Total_Purchases"].sum().reset_index().sort_values(by='Product_Browsing_Time')
L
```

```python
L["Ratio"] = L["Total_Purchases"] / L["Product_Browsing_Time"]
L = L.sort_values('Ratio')
L
```

```python
sns.barplot(data=L,x="Location",y="Ratio")
```

# 19.  Find out the Ratio of browsing time w.r.t total purchases by Device_Type

```python
D_t= cd.groupby(["Device_Type"])["Product_Browsing_Time","Total_Purchases"].sum().reset_index().sort_values(by='Product_Browsing_Time')
D_t
```

```python
D_t["Ratio"] = D_t["Total_Purchases"] / D_t["Product_Browsing_Time"]
D_t= D_t.sort_values('Ratio')
D_t
```

```python
sns.barplot(data=D_t,x="Device_Type",y="Ratio")
```
